```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note: seaborn has some built in data sets

```python
#total bill vs tip data set
tips=sns.load_dataset('tips')
tips.head()
```

```python
#setting a theme for the plots
sns.set_theme()
```

```python
#visualize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
 #load the iris dataset
 iris =sns.load_dataset('iris')
 iris.tail()
```

```python
sns.scatterplot(data=iris,x='sepal_length',y='sepal_width',hue='species')
```

```python
#houseprice dataset
from sklearn.datasets import fetch_california_housing
house_california = fetch_california_housing()#this is a numpy array and to convert this into a dataframe we use pandas lib
house=pd.DataFrame(house_california.data,columns=house_california.feature_names)
house['PRICE']=house_california.target
house.head()
```

```python
print(house_california)
```

distribution plot

```python
sns.distplot(house['PRICE']) #gives the range in which the data is included
```

correlation:



-Positive Correlation



-Negative Correlation

Heat map


```python
correlation=house.corr()
```

```python
#constructing a heat map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
