Matpotlib:


*   useful for making plots



```python
#importing matpotlib library
import matplotlib.pyplot as plt
```

```python
import numpy as np
#to get data for our plots
```

```python
x =np.linspace(0,10,100)
y=np.sin(x) #gives sin values of x
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z) #cosine wave
plt.show()
```

```python
#adding x label and y label to graphs
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
plt.plot(x,z)
plt.xlabel('x')
plt.ylabel('cosine value')
plt.title('cosine wave')
plt.show()
```

```python
#ploting a parabola using matplotlib library
# x=np.linspace(0,10,100)
x=np.linspace(-10,10,100)
y=x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('parabola')
plt.show()
```

```python
plt.plot(x,y,'g+')
plt.show()
```

```python
plt.plot(x,y,'y+')
plt.show()
```

```python
x=np.linspace(-5,5,25)
plt.plot(x,np.sin(x),'r-')
plt.plot(x,np.cos(x),'g--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1]) # 0 and 0 represents origin and 1 represents the height and width
languages =['english','french','spanish','german','hindi']
people=[100,50,150,35,170]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('no of people')
plt.title('bar graph')
plt.show()
```

Pie chart

```python
fig1= plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages =['english','french','spanish','german','hindi']
people=[100,50,150,35,170]
ax.pie(people,labels=languages)
plt.show()
```

```python
 fig1= plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages =['english','french','spanish','german','hindi']
people=[100,50,150,35,170]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

Scatter plot

```python
z=np.linspace(0,10,30)
x=np.sin(z)
y=np.cos(z)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(z,x,color='g')
ax.scatter(z,y,color= 'b')
plt.show()
```

3d scatter plot


```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100) #this wil plot a spiral graph
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
